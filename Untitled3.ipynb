```python
import numpy as np

#Estabelcer la funion sigmoide, se usa como funcion de activacion

def sigmoid(x):
  return 1/ (1 + np.exp(-x))

#Calcular la derivada de la funcion sigmoide
def sigmoid_derivate(x):
  return x * (1 - x)

#Crear la clase para la red neurona
class RedNeuronal:
  #Crear el constructor de la clase
  def __init__(self, n_entradas, n_neuronas_capa_oculta, n_neuronas_capa_salida, tasa_aprendizaje = 0.1):
    self.tasa_aprendizaje = tasa_aprendizaje
    #Inicializando los pesos y las bias para capa oculta
    self.pesos_capa_oculta = np.random.randn( n_entradas, n_neuronas_capa_oculta)
    self.bias_capa_oculta = np.random.randn(n_neuronas_capa_oculta)
    #Inicializando los pesos y las bias para capa salida
    self.pesos_capa_salida = np.random.randn( n_neuronas_capa_oculta, n_neuronas_capa_salida)
    self.bias_capa_salida = np.random.randn(n_neuronas_capa_salida)

  #ir hacia adelante
  def forward(self, entradas):
      #Ir hacia adelante a traves de la capa ocualta
      self.salida_capa_oculta = sigmoid(np.dot(entradas,self.pesos_capa_oculta)+self.bias_capa_oculta)

      #Ir hacia adelante a traves de la capa salida
      self.salida_red = sigmoid(np.dot(self.salida_capa_oculta,self.pesos_capa_salida)+self.bias_capa_salida)

  #Funcion backpropagation
  def backpropagation(self, entradas, salidas_esperadas):
    #Ir hacia adelnte
    self.forward(entradas)

    #Calcular el error de la red
    error = salidas_esperadas - self.salida_red

    #caluclar error cuadratico medio
    error_mse = np.mean(np.square(error))


    #Error en la capa oculta
    error_capa_oculta = np.dot(2*error* sigmoid_derivate(self.salida_red),self.pesos_capa_salida.T)
    d_pesos_oculta = np.dot(entradas.T, error_capa_oculta * sigmoid_derivate(self.salida_capa_oculta))

    d_pesos_salida = np.dot(self.salida_capa_oculta.T, error * sigmoid_derivate(self.salida_red))

    #actualizar pesos y bias
    self.pesos_capa_oculta += self.tasa_aprendizaje * d_pesos_oculta
    self.pesos_capa_salida += self.tasa_aprendizaje * d_pesos_salida
    self.bias_capa_oculta += self.tasa_aprendizaje * np.sum(error_capa_oculta * sigmoid_derivate(self.salida_capa_oculta), axis = 0)
    self.bias_capa_salida += self.tasa_aprendizaje * np.sum(2* error * sigmoid_derivate(self.salida_red), axis = 0)

    return error_mse

red = RedNeuronal(2,4,1, tasa_aprendizaje = 0.1)
# pasar los datos de entrenamiento
entradas = np.array([[0, 0], [0, 1], [1 , 0], [1, 1]])
salidas_esperadas = np.array([[0],[1],[1],[0]])

#entrenamiento

for i in range(10000):
  error_mse = red.backpropagation(entradas, salidas_esperadas)
  print(f'Iteracion {i+1}, Tasa de aprendizaje: {red.tasa_aprendizaje}, Error MSE: {error_mse}')
  print('Pesos de la capa oculta: \n', red.pesos_capa_oculta)
  print('Pesos de la capa de salida: \n', red.pesos_capa_salida)
  print('Bias capa oculta: \n', red.bias_capa_oculta)
  print('Bias capa de salida: \n', red.bias_capa_salida)

#probar la red despues de realizar el entrenamiento

print(red.forward(entradas))




```
